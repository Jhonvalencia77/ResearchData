```python
import numpy as np 
import requests
```

## requests toma el enlace que corresponde a un archivo scv 
### genfromtxt es utilizado para especificar el delimitador, los campos de nombre si se tiene en la primera linea e incluso especificar que los comentarios del csv son "#"

```python
spots = requests.get("http://www.sidc.be/silso/INFO/snmtotcsv.php", stream=True) 

sunspots = np.genfromtxt(spots.raw, delimiter=";") #np.genfromtxt(spots.raw, delimiter=";", names=True, comments="#")
```

```python
sunspots[0][0]
```

## Graficamos con matplotlib

```python
from matplotlib import pyplot as plt

plt.plot(sunspots[:,2], sunspots[:,3]) #sintaxis de np para acceder a todas las filas y columnas 
```

## Si el archivo csv no tiene los nombres en la primera linea, entonces los especificamos mediante una lista. 

```python
spots = requests.get("http://www.sidc.be/silso/INFO/snmtotcsv.php", stream=True)

sunspots = np.genfromtxt(
    spots.raw,
    delimiter=";",
    names=["year", "month", "date", "mean", "deviation", "observations", "definitive"],
    dtype=[int, int, float, float, float, int, int],
) 
```

```python
sunspots["year"]
```

```python
plt.plot(sunspots["year"], sunspots["mean"])
```
