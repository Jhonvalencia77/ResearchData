```python
import json
```

## Creamos un grupo anidado de diccionarios

```python
mydata = {"key": ["value1", "value2"], "key2": {"key4": "value3"}}
```

## Serializamos los datos a string para que puedan ser almacenados en JSON

json.dumps(mydata)

## Observamos los datos con un identaci√≥n 

```python
print(json.dumps(mydata, indent=4))
```

## Creamos un archivo json

```python
%%writefile myfile.json
{"somekey": ["a list", "with values"]}
```

## Abrir un archivo json 

```python
with open("myfile.json", "r") as f:
    mydataasstring = f.read()
```

```python
mydataasstring 
```

## Cargamos el contenido del json a una variable

```python
mydata = json.loads(mydataasstring) 
```
